```python
import pandas as pd
import numpy as np
```

```python
X_df=pd.read_csv("train_feature.csv")
y_df=pd.read_csv("train_label.csv")
X_df
```

## 1.数据预处理

### 异常值处理（分箱）

```python
lst = []
cols = X_df.columns
for col in cols:
    df2_col = X_df[col]
    q1 = df2_col.quantile(0.1)
    q3 = df2_col.quantile(0.9)
    iqr = q3-q1
    ma = q3+iqr*5
    mi = q1-iqr*5  #这里的范围设的较大，清除严重离群数据
    error = df2_col[(df2_col>ma) | (df2_col<mi)]   # 筛出异常
    df2_col[(df2_col>ma) | (df2_col<mi)]=np.nan
    lst.append(error)
    err_data = pd.concat(lst)
print('一共检测到异常值共：%i'%len(err_data),'\n')
X_df.describe()
```

### 单位化

```python
def minmax_norm(df):
    return (df - df.min()) / ( df.max() - df.min())

X_df=minmax_norm(X_df)
X_data=X_df.to_numpy()
y_data=y_df.to_numpy().squeeze()
X_df.describe()
```

### 空值填充（Knn）

```python
from sklearn.impute import KNNImputer
imputer = KNNImputer(n_neighbors=3, weights="uniform")
X_data=imputer.fit_transform(X_data)
```

### 特征选择

```python
from sklearn.neighbors import NearestNeighbors
class ReliefF(object):
    def __init__(self,best_features_num=10):
        self.best_features = None
        self.sorted_feature = None
        self.statistics = None
        self.features_num = None
        self.best_features_num= best_features_num
    def get_statistics(self,X,y):
        #假设已经规范化处理
        self.features_num=X.shape[1]
        labels,num=np.unique(y,return_counts=True)
        statistics=np.zeros(self.features_num)
        sample_num=X.shape[0]
        for j in range(self.features_num):
            for i in range(sample_num):
                #find nearest neighbor
                nearest_norms=self.features_num*np.ones(labels.size)
                nearest_ind= np.zeros(labels.size)
                for k in range(sample_num):
                    label_index=np.where(labels==y[k])
                    temp_norm=np.linalg.norm(X[i]-X[k])
                    if nearest_norms[label_index] > temp_norm > 0:
                        nearest_norms[label_index]=temp_norm
                        nearest_ind[label_index]=k
                for count in range(labels.size):
                    if labels[count]==y[i]:
                        statistics[j]+=-(X[i][j]-X[int(nearest_ind[count])][j])**2
                    else:
                        statistics[j]+=num[count]/sample_num*(X[i][j]-X[int(nearest_ind[count])][j])**2
        sorted_index=np.argsort(-statistics)
        self.sorted_feature=sorted_index
        self.statistics=statistics[sorted_index]
        self.best_features=sorted_index[:self.best_features_num]

#因为数据集较大,可以多次随机取样用 RelifF 方法选择较好的特征,
# 最后从多次采样中选择出现次数多的好特征
itr_times=10
arr=np.array([])
for i in range(itr_times):
    np.random.seed(43*i)
    print(i)
    slice=np.random.choice(X_data.shape[0],150)
    X_features=X_data[slice]
    y_features=y_data[slice]
    relieff=ReliefF(25)
    relieff.get_statistics(X_features,y_features)
    arr = np.append(arr,relieff.best_features)
features,num=np.unique(arr,return_counts=True)
features=features.astype(int)
sorted_index=np.argsort(-num)
sorted_features = features[sorted_index]
print("final result:")
print(sorted_features)
print(num[sorted_index])
best_feature=features[np.where(num>=7)]
print(best_feature)
```

```python
best_feature=[84,53,114,64,73,101,37,9] #这个是多次实验选择出的较好的特征
# best_feature=[  5  ,37  ,43  ,47  ,50  ,64  ,67  ,84  ,92 , 99, 104 ,114]
# best_feature=[ 73, 68 , 40 , 37 , 42 , 82 ,118 ,112 , 23 , 32 , 48 , 99 , 29 , 38 ,  8 , 72 , 44 ,119]
# best_feature= [  5  ,37  ,46  ,53  ,67  ,84 ,104]
X_test=X_data[:,best_feature]
X_test.shape
```

### 数据集划分

```python
from sklearn.model_selection import train_test_split
training_rate=0.7
validation_rate=0.3
data_size = y_data.size
X_training_data,X_validation_data,y_training_data,y_validation_data = train_test_split(X_test,y_data,test_size=0.30,random_state=7)



labels=np.unique(y_training_data)
label_num=labels.size
```

## 模型训练

### 决策树模型

```python
from sklearn import tree
clf = tree.DecisionTreeClassifier(criterion='entropy', max_depth=5, min_samples_leaf=31,
                       min_samples_split=11, random_state=21,
                       splitter='random')
clf = clf.fit(X_training_data,y_training_data)
result = clf.score(X_validation_data,y_validation_data)
print(result)
clf.predict(X_validation_data)
```

## 参数调试

```python
import numpy as np
from sklearn.model_selection import GridSearchCV
from sklearn.model_selection import RandomizedSearchCV
# 参数搜索空间
param_grid = {
    'max_depth': np.arange(1, 10, 1),
    'random_state': np.arange(1, 50, 10),
    'min_samples_leaf': np.arange(1, 100, 10),
    'min_samples_split': np.arange(1, 50, 10)

}
# 网格搜索模型参数
grid_search = GridSearchCV(clf, param_grid, cv=5)#, scoring='f1_micro')
#cv:k折交叉验证参数
grid_search.fit(X_test, y_data)
print(grid_search.best_params_)
print(grid_search.best_score_)
print(grid_search.best_estimator_)
```
